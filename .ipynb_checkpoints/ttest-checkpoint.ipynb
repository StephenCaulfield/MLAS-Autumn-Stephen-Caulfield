<h2 align = "center">1. PERFORMING PAIRED SAMPLES T-TEST WITH SCIPY</h2>

<h3>WHAT IS A T-TEST</h3>
<p>T-test is a statistical test comparing the means of two different samples. The specific type of T-test we are performing is a paired sample T-test, in which the samples have values that correspond with eachother(E.g. A class of students performance on exams before studying and after studying)</p>

t=m−μs/n−−√

We first need to import the necessary packages.

```python
import numpy as np
import statsmodels.stats.weightstats as stat
import scipy.stats as ss
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
m_a, s_a, s_b = 1.0, 0.4, 0.4
m_b = m_a + 2
N = 40

a = np.random.normal(loc=m_a, scale=s_a, size=N)
b = np.random.normal(loc=m_b, scale=s_b, size=N)


ss.ttest_rel(a, b)
```

```python
df = pd.DataFrame({'Category': ['A'] * len(a) + ['B'] * len(b), 'Value': np.hstack([a,b])})


sns.catplot(x='Category', y='Value', jitter=False, data=df);
```

```python
min_x = min(m_a, m_b) - 5.0 * max(s_a, s_b)
max_x = max(m_a, m_b) + 5.0 * max(s_a, s_b)
x = np.linspace(min_x, max_x, 1000)
y_a = ss.norm.pdf(x, m_a, s_a)
y_b = ss.norm.pdf(x, m_b, s_b)
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x, y_a)
ax.plot(x, y_b)
plt.show()
```

<h2 align = "center">3. PERFORMING A T-TEST ON THE DATA SET</h2>

```python
#Break down of paired t-test formula: https://www.youtube.com/watch?v=_7IW2PUqe64
def t_val(dfi_pre, dfi_post):
    n = dfi_pre.count()
    dof = n - 1
    tot_arr = []
    v = 0
    for i in range(len(dfi_pre)):
        v = dfi_pre[i] - dfi_post[i]
        tot_arr.append(v)
    m_x = sum(tot_arr)/n
    s=0
    for i in range(len(tot_arr)):
        s += (dof - tot_arr[i]) ** 2
    s = s/(n-1)
    s = math.sqrt(s)
    se = s/math.sqrt(n)
    t = (m_x - 0)/(se/math.sqrt(n))
    return t
```

```python
t1 = t_val(dfdiet1['pre.weight'], dfdiet1['weight6weeks'])
t1
```

```python
t2 = t_val(dfdiet2['pre.weight'], dfdiet2['weight6weeks'])
t2
```

```python
t3 = t_val(dfdiet3['pre.weight'], dfdiet3['weight6weeks'])
t3
```
